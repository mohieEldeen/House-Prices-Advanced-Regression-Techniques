```python
#dealing with the to be one hotted features of the train data

#the features which will be dropped to avoid over fitting
tobedroped=[]

#a function to creat the one hotted features
def getonehot(feature , lis , sw):
    
    
    global tobedroped
    
    columns=feature.unique() #creating a list of the new one hotted features
    
    new_onehot=pd.get_dummies(feature,prefix=feature.name) #creating the one hotted features
    
    #this loop is to check if there're feature included in the lis but not in the new one hotted features
    #this is done because sometimes there're different kinds of catagories of the same feature in the train and test set
    #so if there're other features that's included is the train but not in the test, it will be added as 0  
    #and that's done to have the same number of features for the training and test sets
    for i in lis :
        if i not in columns :
            nam=feature.name+'_'+str(i)
            new_onehot[nam]=0
    if sw:
        tobedroped.append(new_onehot.iloc[:,0].name) # adding the first one hotted feature to the tobedroped list
    return new_onehot


train_hot=pd.DataFrame() #new data frame for the one hotted features in train set

#creating the one hotted features for training
for j in  tobeonehoted:   
    train_hot=pd.concat([train_hot,getonehot(train[j],test[j].unique(),True)],axis=1)
```

```python
test_hot=pd.DataFrame() #new data frame for the one hotted features in test set

#creating the one hotted features for test sets
for j in  tobeonehoted:   
    test_hot=pd.concat([test_hot,getonehot(test[j],train[j].unique(),False)],axis=1)
```
